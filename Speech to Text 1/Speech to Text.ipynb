Task 2 & 3. Analyze that audio using librosa waveplot and spectrogram and Comment on the Output

```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
```

```python
audio_path="A1a.mp3"
x,sr=librosa.load(audio_path)
```

```python
ipd.Audio(audio_path)
```

```python
plt.figure(figsize=(14,5))
librosa.display.waveshow(x,sr=sr)
```

```python
X=librosa.stft(x)
Xdb=librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14,5))
librosa.display.specshow(Xdb,sr=sr,x_axis='time',y_axis='hz')
plt.colorbar()
```

1. We observe that every word has a distinct presence in the Wave chart
2. We also observe that distinct parts of the words show a specific presence.
